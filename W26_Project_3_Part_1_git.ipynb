````python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Levenshtein
from sklearn.preprocessing import StandardScaler

path = kagglehub.dataset_download("berkeleyearth/climate-change-earth-surface-temperature-data")
df = pd.read_csv(filepath_or_buffer=path+'/GlobalLandTemperaturesByCountry.csv')
````

````python
# Part1.1
# Filter the rows from 1901-01-01 to 2012-12-31.
df = df[(df['dt'] >= '1901-01-01') & (df['dt'] <= '2012-12-31')]
# Create a dataframe showing the AverageTemperature with the countries as row and months from January 1901 to December 2012 as columns.
df_monthly = df.pivot(index='Country', columns='dt', values='AverageTemperature')
# Remove the rows with missing values and display the resulting dataframe.
df_monthly = df_monthly.dropna()
# Combine the columns by year and display the resulting dataframe.
df_monthly.columns = pd.to_datetime(df_monthly.columns) # Convert the column names to datetime objects
years = df_monthly.columns.year # Extract the year from the datetime objects
df_yearly = df_monthly.T.groupby(years).mean().T
print(f"Part 1.1: Shape of df_yearly: {df_yearly.shape}")
# Part1.2
def plot_country_time_series(df, countries):
    country_num = len(countries)
    fig, ax = plt.subplots(country_num, 1, figsize=(12, 5 * country_num))

    # If there's only one country, ax will not be an array, so convert it to a list for consistent iteration
    if country_num == 1:
        ax = [ax]

    for i, country in enumerate(countries):
        if country in df.index:
            ax[i].plot(df.columns, df.loc[country].values)
            ax[i].set_title(country)
            ax[i].set_ylabel('Avg. Temperature')
            ax[i].set_xlabel('Year')

    plt.show()
print("\nPart 1.2: ")
plot_country_time_series(df_yearly, ['Zambia'])

# Part1.3: Find the clostest 3 Countries by using Levenshtein distance.
countries_dis = []
for country in df_yearly.index:
  dis = Levenshtein.distance('Haitao', country)
  countries_dis.append((country, dis))
countries_dis.sort(key=lambda x: x[1])
clostest_3_countries = countries_dis[:3]
print(f"Part 1.3: \nClosest 3 countries to 'Haitao': {clostest_3_countries}")

# Part1.4: Standardize the monthly temperature data.
scaler = StandardScaler()
monthly_stand_lst = scaler.fit_transform(df_monthly)

# Create a new DataFrame with the standardized values
df_monthly_stand = pd.DataFrame(
    monthly_stand_lst, 
    index=df_monthly.index, 
    columns=df_monthly.columns
)
print(f"\nPart 1.4: \n{df_monthly_stand.head()}")

# Part 1.5: Run the function in step 1 on both the original dataset as well as the standardized dataset
selected_countries = [x[0] for x in clostest_3_countries] + ['Canada', 'Oman', 'Belgium']
print(f"\nPart 1.5: \n")
plot_country_time_series(df_monthly, selected_countries)
plot_country_time_series(df_monthly_stand, selected_countries)
````
